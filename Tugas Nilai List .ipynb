Diberikan sekumpulan Nilai Random antara 0 sampai dengan 100 sebanyak 250 nilai. Buatlah program untuk melakukan hal berikut:

Tampilkan Angka Ganjil dari ke 250 nilai tersebut

Tampilkan berapa banyak yang mendapatkan nilai lebih dari 70

Tampilkan standar deviasi dari sekumpulan nilai tersebut

Tampilkan nilai maksimum dan minimum

Tampilkan modus dari sekumpulan nilai tersebut

```python
import random

# GENERATE RANDOM
nilaiTugas = []
for i in range(250):
    nilaiTugas.append(random.randrange(1,100,1))

print(nilaiTugas)
```

```python
# 1 Mencari Nilai Ganjil
for n in nilaiTugas:
    if (n % 2 == 1):
        print (n,"ganjil")
```

```python
# 2 Tampilkan berapa banyak yang mendapatkan nilai lebih dari 70
counter = 0
for x in nilaiTugas:
    if x>=70:
        counter +=1
        print (x)
        print ("jumlah >=70 : ",counter)
```

```python
# Tampilkan standar deviasi dari sekumpulan nilai tersebut
jumlah = 0
for i in range(len(nilaiTugas)):
    jumlah += nilaiTugas [i]

rata = jumlah/(len(nilaiTugas))
print ("Nilai rata-rata :",rata)

sigma = 0
for i in range(len(nilaiTugas)):
    hitung =(nilaiTugas [i]-rata)**2
    sigma += hitung

pembagianN = sigma/(len(nilaiTugas)-1)
SD = pembagianN**0.5
print ("Nilai Standar Deviasi:",SD)
```

```python
#Tampilkan nilai maksimum dan minimum

#Nilai Maksimum

maks = 0
for ma in nilaiTugas:
    if ma > maks:
        maks=ma
print ("Nilai Max :",maks)

#Nilai Minimum

minimum = 100
for mi in nilaiTugas:
    if mi < minimum:
        minimum=mi
print ("Nilai Min :",minimum)
```

```python
#Menghitung Modus

nilai_modus=nilaiTugas [0]
jumlah=nilaiTugas.count(nilai_modus)
for n in nilaiTugas:
    jumlah_n=nilaiTugas.count (n)
    if jumlah_n>jumlah:
        nilai_modus=n
        jumlah=jumlah_n
print (nilai_modus)
```
